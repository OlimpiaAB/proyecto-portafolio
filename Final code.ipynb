```python
#Creamos un diccionario que identifique y cambie la variable "sex" con (F) y (M) por `Female`y `Male`, respectivamente. 
d={'F':'Female', 'M':'Male'}
math['sex'] = math['sex'].apply(lambda x:d[x])
math['sex'].head() #Un nuevo .head() para comprobar el cambio realizado
```

```python
# ¿Cuántas chicas hay?¿Cuántos chicos hay?
#.value_counts() => Nos ofrece el recuento de valores únicos sobre la variable indicada 
# =============================
math['sex'].value_counts()
```

```python
# Para un resultado más visual introducimos un gráfico de barras
# =============================
ax=sns.barplot(x=math['sex'].value_counts().index,y=math['sex'].value_counts().values,palette="hls")
for i in ax.containers: ax.bar_label(i, label_type='center',color='white', fontsize=18)
plt.xlabel('Sexo', size=14, color='#4f4e4e')
plt.yticks([], [])
plt.title('Número estudiantes por sexo',size=18, color='#4f4e4e')
sns.despine(left = True)
plt.show()
ax.get_figure().savefig('Número estudiantes por sexo.png')
```

```python
#.groupby() => Permite realizar operaciones mediante la agrupación de variables, en nuestro caso por sexo
#.groupby().mean() => Al añadir .mean() obtenemos la media de los grupos, excluyendo los valores faltantes
# =============================
math.groupby('sex').mean()
```

```python
# Con el objetivo de visualizar las calificaciones en su conjunto creamos una nueva variable a la que llamaremos "G"
# =============================
math['GGG']=math['G3']+math['G2']+math['G1']
math['G']=math['GGG']/3
```

```python
#Agrupación por sexo y omisión de la variable creada para el cálculo de la media de las calificaciones
math_GGG=math.set_index('GGG')
groups=math_GGG.groupby('sex')
math_girl=groups.get_group('Female')
math_boy=groups.get_group('Male')
```

```python
#Mapa de calor - Nivel de correlación
plt.figure(figsize=(20,12))
upp_mat = np.triu(math_girl.corr())
ax=sns.heatmap(math_girl.corr(),annot=True,mask = upp_mat,linewidths=1, linecolor='white',cmap='OrRd')
plt.title('Nivel de correlación de las variables para las estudiantas', fontsize = 20)
ax.set_xticklabels(ax.get_xmajorticklabels(), fontsize = 13,rotation=45)
ax.set_yticklabels(ax.get_ymajorticklabels(), fontsize = 13)
plt.show()
ax.get_figure().savefig('chicas.png')
```

```python
plt.figure(figsize=(20,12))
upp_mat = np.triu(math_girl.corr())
ax=sns.heatmap(math_boy.corr(),annot=True,mask = upp_mat,linewidths=1, linecolor='white',cmap='GnBu')
plt.title('Nivel de correlación de las variables para los estudiantes', fontsize = 20)
ax.set_xticklabels(ax.get_xmajorticklabels(), fontsize = 13,rotation=45)
ax.set_yticklabels(ax.get_ymajorticklabels(), fontsize = 13)
plt.show()
ax.get_figure().savefig('chicos.png')
```
